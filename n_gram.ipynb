```python
s = "i am smriti i am learning nlp i am loving it nlp is very interesting i am enjoying it i am loving it"
```

```python
from nltk.tokenize import word_tokenize
```

```python
w = word_tokenize(s)
```

```python
from nltk.collocations import BigramCollocationFinder, TrigramCollocationFinder, ngrams
```

```python
b = BigramCollocationFinder.from_words(w)
b.ngram_fd


```

```python
t = TrigramCollocationFinder.from_words(w)
t.ngram_fd
```

## count vectorizer

```python
l = ["my name is smriti and i am learning nlp", " nlp is very interesting and i am loving it", "i am enjoying learning nlp"]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(l, columns=["name"])
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
new_data = cv.fit_transform(df["name"])
new_data.toarray()
```

```python
cv.vocabulary_
```
